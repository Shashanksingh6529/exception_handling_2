Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.

```python
"When creating a custom exception in Python, it's best practice to inherit from the built-in Exception class. This is because the Exception class serves as the base class for all built-in exceptions in Python. By inheriting from the Exception class, our custom exception can benefit from the features and behaviors defined in the base class, such as the ability to catch it using a generic except clause or to handle it along with other built-in exceptions."

"Additionally, inheriting from the Exception class allows our custom exception to be compatible with existing exception handling mechanisms in Python, making it easier for other developers to understand and work with our code."
```

Q2. Write a Python program to print Python Exception Hierarchy.

```python
def print_exception_hierarchy(exception_class, indent=0):
    print(' ' * indent + str(exception_class))
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)

print_exception_hierarchy(Exception)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

```python
"ImportError: It is raised when the import statement fails to find the module definition or when an imported module fails to load or execute."

"ModuleNotFoundError: It is a subclass of ImportError that is raised specifically when a module is not found during an import operation."
```

Q6. List down some best practices for exception handling in Python.

```python
Be specific with exception handling: Catch specific exceptions rather than using a generic except clause.
Use try-except-else blocks: Use else block to handle code that should only run if no exceptions were raised.
Clean up with finally: Use finally block for cleanup code that should always run, regardless of whether an exception occurred or not.
Avoid catching Exception: Avoid catching the base Exception class unless absolutely necessary, as it may catch unexpected exceptions."
Log exceptions: Use logging to record exceptions for debugging and monitoring purposes."
Keep exception messages informative: Provide meaningful error messages to aid in troubleshooting."
Handle exceptions close to the source: Catch exceptions as close to the source of the error as possible for better error localization and debugging."
```
